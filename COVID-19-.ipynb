```python
# we need to clean Pulmonary score from expression to string to float64

# Importing re package for using regular expressions 
import re 
  
# Function to clean the names 
def Clean_names(name): 
    # Search for opening bracket in the name followed by 
    # any characters repeated any number of times 
    if re.search('<', name): 
  
        # Extract the position of beginning of pattern 
        pos = re.search('<', name).start() 
  
        # return the cleaned name 
        return name[pos+1:] 
  
    else: 
        # if clean up needed return the same name 
        return name 
          
# Updated the city columns 
dataset_train['Pulmonary score'] = dataset_train['Pulmonary score'].apply(Clean_names) 
dataset_test['Pulmonary score'] = dataset_test['Pulmonary score'].apply(Clean_names) 
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dropout, Dense
from keras.optimizers import Adam,Nadam
def ann_model():
    model = Sequential()
    model.add(Dense(64, activation="relu",kernel_initializer = 'uniform', input_dim=32))
    model.add(Dropout(0.1))
    model.add(Dense(activation="relu", kernel_initializer = 'uniform',units=256))
    model.add(Dropout(0.1))
    model.add(Dense(activation="relu", kernel_initializer = 'uniform',units=256))
    model.add(Dropout(0.1))
    model.add(Dense(activation="relu", kernel_initializer = 'uniform',units=256))
    model.add(Dropout(0.1))
    model.add(Dense(activation="relu", kernel_initializer = 'uniform',units=128))
    model.add(Dropout(0.1))
    model.add(Dense(activation="relu", kernel_initializer = 'uniform',units=128))
    model.add(Dropout(0.1))
    model.add(Dense(activation="relu", kernel_initializer = 'uniform',units=128))
    model.add(Dropout(0.1))
    model.add(Dense(activation="relu", kernel_initializer = 'uniform',units=64))
    
    model.add(Dense(activation="sigmoid", units=1))
    model.compile(optimizer = 'adam',loss = 'mse', metrics = [ 'mae'])
    return model

model = ann_model()
model.summary()

# try:
#     tbCallbacks = keras.callbacks.TensorBoard(log_dir='./logs_2', histogram_freq=0, batch_size=128, write_graph=True, write_grads=True, write_images=True, embeddings_freq=0, embeddings_layer_names=None, embeddings_metadata=None, embeddings_data=None, update_freq='epoch')
#     model.fit(X_train_temp, Y_train_temp, epochs=100,batch_size=32)
# except KeyboardInterrupt:
#     print("\nInterrupting")
```
